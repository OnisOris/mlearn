$\hat{y} = f(x)$ - функция соответсвия или прогностическая функция или прогнозная функция
Между признаками и откликами ищется в некотором классе функций $f_w(\textbf{x})$ параметризованном вектором параметров **w** - вектор параметров (веса модели) (почему тут столько названий у одного определения)

Класс функций может быть множеством всех константных прогнозов: 
$$f_w(\textbf{x}) = w, w \in \mathbb{R}$$

Или состоять из всех линейных функций признаков

$$f_w(x) = w_0 + w_1x^1 + w_2x^2+...+w_Dx^D, \textbf{w} = [w_0, w_1,...w_D]$$

```python
def f_w(x: float, weights: np.ndarray) -> float:
    f_w_out = 0.
    for i in range(weights.size):
        f_w_out += weights[i] * x^i
    return f_w_out
```

```python
f_w(a[0], b)
```

```python
def f_w_v(x_v: np.ndarray, weights: np.ndarray) -> np.ndarray:
    x_extend = np.hstack([[1], x_v])
    out = np.array([])
    for x in x_v:
        out = np.hstack([out, f_w(x, weights)])
    return out
```

```python
x_arr = np.array([2, 1, 3, 4])
w_arr = np.array([3, 2, 4, 5])
```

```python
plt.plot(w_arr, f_w_v(x_arr, w_arr))
```
